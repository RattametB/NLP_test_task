```python
import Pdf2text
```

```python
file_path = "C:/Users/LENOVO/Downloads/LakeIV_B-1-50.pdf" #your file path
```

```python
detected_text = Pdf2text.text_extracter(file_path,n_page = 10)
```

```python
detected_text
```

```python
cleaned_text = Pdf2text.clean_text(detected_text)
```

```python
cleaned_text
```

```python
rus,eng = Pdf2text.language_saperator(cleaned_text)
```

```python
rus
```

```python
eng
```

```python
tokens_rus = Pdf2text.tokenize_text(rus,'russian')
```

```python
tokens_rus[0:30]
```

```python
tokens_eng = Pdf2text.tokenize_text(eng,'english')
```

```python
tokens_eng[0:30]
```

```python
tokens_rus_nsw = Pdf2text.remove_stopwords(tokens_rus,'russian')
```

```python
tokens_rus_nsw[0:30]
```

```python
tokens_eng_nsw = Pdf2text.remove_stopwords(tokens_eng,'english')
```

```python
tokens_eng_nsw[0:30]
```

```python
lemma_rus = Pdf2text.lemmatize_text(tokens_rus_nsw,'russian')
```

```python
lemma_rus[0:30]
```

```python
lemma_eng = Pdf2text.lemmatize_text(tokens_eng_nsw,'english')
```
